```python
import pandas as pd
import math
from random import randint

Data = pd.read_csv('D://DataTrain.csv')
Data
```

```python
no = []
att1 = []
att2 = []
att3 = []
att4 = []
label = []

for i in range(0, len(Data)):
    pointer = randint(0, len(Data))
    no.append("B" + str(i + 1))
    att1.append(Data.Att_1[pointer])
    att2.append(Data.Att_2[pointer])
    att3.append(Data.Att_3[pointer])
    att4.append(Data.Att_4[pointer])
    label.append(Data.Label[pointer])
    
for i in range(0, len(Data)):
    s = no[i] + " - " + str(att1[i]) + " - " + str(att2[i]) + " - " + str(att3[i]) + " - " + str(att4[i]) + " - " + str(label[i])
    print(s)
```

```python
satu = 0
nol = 0
for i in range(0, len(Data)):
    if(Data.Label[i] == 1):
        satu += 1
    else:
        nol += 1
        
print("Satu - " + str(satu))
print("Nol - " + str (nol))
```

```python
satu = 0
nol = 0
for i in range(0, len(Data)):
    if(label[i] == 1):
        satu += 1
    else:
        nol += 1
        
print("Satu - " + str(satu))
print("Nol - " + str (nol))
```

```python
raw_data = {'no': no,
        'att1': att1,
        'att2': att2,
        'att3': att3,
        'att4': att4,
        'label': label}

df = pd.DataFrame(raw_data, columns = ['no', 'att1', 'att2', 'att3', 'att4', 'label'])

df.to_csv('D://RandomTrain.csv', index=False)
```
