```python
import pandas as pd
import math
from random import randint

Data = pd.read_csv('D://RandomTrain.csv')
Data
```

```python
avg_att1 = 0
avg_att2 = 0
avg_att3 = 0
avg_att4 = 0

for i in range(0, len(Data)):
    avg_att1 += Data.att1[i]
    avg_att2 += Data.att2[i]
    avg_att3 += Data.att3[i]
    avg_att4 += Data.att4[i]
    
avg_att1 = avg_att1 / len(Data)
avg_att2 = avg_att2 / len(Data)
avg_att3 = avg_att3 / len(Data)
avg_att4 = avg_att4 / len(Data)

print("AVG Att 1: " + str(round(avg_att1, 2)))
print("AVG Att 2: " + str(round(avg_att2, 2)))
print("AVG Att 3: " + str(round(avg_att3, 2)))
print("AVG Att 4: " + str(round(avg_att4, 2)))
```

```python
att1_true_yes = 0;
att1_true_no = 0;
att1_false_yes = 0;
att1_false_no = 0;

for i in range(0, len(Data)):
    if(Data.att1[i] >= avg_att1 and Data.label[i] == 1):
        att1_true_yes += 1
    elif(Data.att1[i] >= avg_att1 and Data.label[i] == 0):
        att1_true_no += 1
    elif(Data.att1[i] < avg_att1 and Data.label[i] == 1):
        att1_false_yes += 1
    elif(Data.att1[i] < avg_att1 and Data.label[i] == 0):
        att1_false_no += 1

print("Att 1")
print("True and Yes: " + str(att1_true_yes))
print("True and No: " + str(att1_true_no))
print("False and Yes: " + str(att1_false_yes))
print("False and No: " + str(att1_false_no))

total_att1_true = att1_true_yes + att1_true_no
total_att1_false = att1_false_yes + att1_false_no
total_att1 = total_att1_true + total_att1_false

if(total_att1_true != 0):
    gini_att1_true = 1 - ((att1_true_yes/total_att1_true)**2) - ((att1_true_no/total_att1_true)**2)
else:
    gini_att1_true = 0
if(total_att1_false != 0):
    gini_att1_false = 1 - ((att1_false_yes/total_att1_false)**2) - ((att1_false_no/total_att1_false)**2)
else:
    gini_att1_false = 0
gini_att1 = ((total_att1_true/total_att1)*gini_att1_true) + ((total_att1_false/total_att1)*gini_att1_false)

print("Att1 true        : " + str(total_att1_true))
print("Att1 false       : " + str(total_att1_false))
print("Total Att1       : " + str(total_att1))
print("Gini(Att1_true)  : " + str(round(gini_att1_true, 3)))
print("Gini(Att1_false) : " + str(round(gini_att1_false, 3)))
print("Gini(Att1)      : " + str(round(gini_att1, 3)))
```
